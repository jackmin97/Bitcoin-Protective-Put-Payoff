# 1. Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
```

# 2. Put payoff 

```python
def put_payoff(St, strike_price, premium):
    return np.where(St < strike_price, strike_price - St, 0) - premium
```

# 3. Define parameters

```python
# Bitcoin price 
spot_price = 17500 

# Bitcoin purchase price
bitcoin_purchase_price = 17500 

# Long put
strike_price_long_put = 17500
premium_long_put = 500

# Bitcoin range at expiration of the put
St = np.arange(0.9*spot_price,1.1*spot_price,1) 
```

# 4. Long 17500 strike put payoff

```python
payoff_long_put = put_payoff(St, strike_price_long_put, premium_long_put)
# Plot
fig, ax = plt.subplots(figsize=(8,5))
ax.spines['bottom'].set_position('zero')
ax.plot(St,payoff_long_put,label='Long 17500 Strike Put',color='r')
plt.xlabel('Bitcoin Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

# 5. Long Bitcoin payoff

```python
payoff_bitcoin = St - bitcoin_purchase_price
# Plot
fig, ax = plt.subplots(figsize=(8,5))
ax.spines['bottom'].set_position('zero')
ax.plot(St,payoff_bitcoin,label='Long Bitcoin',color='g')
plt.xlabel('Bitcoin Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

# 6. Protective put payoff

```python
payoff_protective_put = payoff_bitcoin + payoff_long_put

# Plot
fig, ax = plt.subplots(figsize=(8,5))
ax.spines['bottom'].set_position('zero')
ax.plot(St,payoff_bitcoin,'--',label='Long Bitcoin',color='g')
ax.plot(St,payoff_long_put,'--',label='Long 17500 Strike Put',color='r')
ax.plot(St,payoff_protective_put,label='Protective Put')
plt.xlabel('Bitcoin Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```
